```python
import zipfile, send2trash, time, shutil, os
from pathlib import Path
from os import remove
from shutil import move
```

```python
import logging
logfile = Path('C:/Users/***/Archive/Logs/Logfile.txt') # set to variable later
logging.basicConfig(filename=str(logfile),
                    filemode='a',
                    format='%(asctime)s,%(msecs)03d %(name)s %(levelname)s %(message)s',
                    datefmt='%Y-%m-%d %H:%M:%S',
                    level=logging.DEBUG, force=True)
```

```python
logging.info("Running Unzip_files")
```

```python
folder = Path('C:/Users/***/csv/') # set to variable later
mainfolder = Path('C:/Users/***/') # set to variable later

WEEKLY = f'{folder}\\WEEKLY.csv'
MONTHLY = f'{folder}\\MONTHLY.csv'
price = f'{folder}\\price.csv'
```

```python
ZF = list(mainfolder.glob('*.zip'))
```

```python
time.sleep(1)
```

```python
for file in ZF:
    with zipfile.ZipFile(file, 'r') as zip_ref:
        # Extract all the contents into the specified directory
        zip_ref.extractall(folder)
        
```

```python
time.sleep(5)
```

```python
CSV_F = list(folder.glob('*.csv'))
```

```python
TF = list(folder.glob('*.csv'))
for t in TF:
    if 'WEEKLY.csv' in str(t):
        remove(t)
    elif 'MONTHLY.csv' in str(t):
        remove(t)
        
```

```python
time.sleep(5)
```

```python
for f in CSV_F:
    if 'WEEKLY' in str(f):
        os.rename(f,WEEKLY)
    elif 'MONTHLY' in str(f):
        os.rename(f,MONTHLY)
```

```python
time.sleep(5)
```

```python
for file in ZF:
    send2trash.send2trash(file)
```

```python
logging.info("Unzip_files finished successfully")
```
