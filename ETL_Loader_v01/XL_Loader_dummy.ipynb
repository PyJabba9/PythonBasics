```python
import urllib.request, re, pandas as pd
from pathlib import Path
```

```python
folder = Path('C:/Users/***/FTP/') # move to json
```

```python
XL_file = list(folder.glob('*.xlsx'))
```

```python
for xl in XL_file:
    excel = pd.read_excel(xl, usecols=[1,2], header=None, names=['size','name'])
    excel['size'] = excel['size'].str.replace('[','')
    excel['size'] = excel['size'].str.replace(']','')
    excel['size'] = excel['size'].str.replace("'",'')
    excel['size'] = excel['size'].str.replace("'",'')
    excel['name'] = excel['name'].str.replace('[','')
    excel['name'] = excel['name'].str.replace(']','')
    excel['name'] = excel['name'].str.replace("'",'')
    excel['name'] = excel['name'].str.replace("'",'')
    excel = excel.drop([0])
    excel = excel.reset_index()
    excel = excel.drop(columns=['index'])
    if 'M.xlsx' in str(xl):
        excel_m = excel
    elif 'W.xlsx' in str(xl):
        excel_w  = excel
    elif 'price.xlsx' in str(xl):
        excel_p  = excel
```
